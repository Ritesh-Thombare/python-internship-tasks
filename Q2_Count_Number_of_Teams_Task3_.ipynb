Q2 Count Number of Teams

```python
class Solution:
    def numTeams(self, rating):
        n = len(rating)
        ans = 0

        for j in range(n):
            ls = lg = rs = rg = 0

            for i in range(j):
                if rating[i] < rating[j]:
                    ls += 1
                elif rating[i] > rating[j]:
                    lg += 1

            for k in range(j + 1, n):
                if rating[k] > rating[j]:
                    rg += 1
                elif rating[k] < rating[j]:
                    rs += 1

            ans += ls * rg + lg * rs

        return ans
```
